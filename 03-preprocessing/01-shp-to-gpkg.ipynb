```python
from nfw_project import config_paths as cfg

```

```python
from pathlib import Path
import geopandas as gpd
import pandas as pd
```

# National Forest System Roads

```python
# Input file
src_path = cfg.NFS_ROADS_SHP
# Read roads
roads = gpd.read_file(src_path)
print("Rows:", len(roads), "CRS:", roads.crs)
```

```python
# Write to GeoPackage as layer "nfs_roads"
roads.to_file(cfg.USFS_GPKG, layer="nfs_roads", driver="GPKG")

print("Wrote layer 'nfs_roads' to:", cfg.USFS_GPKG)
```
